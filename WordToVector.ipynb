```python
from gensim.models import Word2Vec
import gensim
#used for topic modeling, document similarity analysis
from nltk.tokenize import sent_tokenize, word_tokenize
#used for tokenizing sentences and words, respectively, in natural language processing tasks.
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
sample = open("/content/sample_data/alice.txt")
s = sample.read()
f = s.replace("\n"," ")
f
```

```python
import nltk
nltk.download('punkt')

# iterate through each sentence in the file
data = []
for i in sent_tokenize(f):
  temp = []     # empty list

  # tokenize the sentence into words
  for j in word_tokenize(i):
    temp.append(j.lower())

  data.append(temp)

print(data)
```

**Word Embedding**

```python
# Create CBOW model
model1 = gensim.models.Word2Vec(data, min_count=1, vector_size=100, window= 5)
```

```python
print("Cosine similarity between 'alice' "+ "and 'rabbit' - CBOW " , model1.wv.similarity('alice','rabbit'))
```

```python
print("Cosine similarity between 'alice' " + "and 'her' - CBOW : ",
      model1.wv.similarity('alice', 'her'))
```

```python
# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count=1, vector_size=100,
                                window=5, sg=1)
```

```python
# Print results
print("Cosine similarity between 'alice' " +
      "and 'rabbit' - Skip Gram : ",
      model2.wv.similarity('alice', 'rabbit'))
```
